```python
# Clean PDF and store metadata files
import pdfplumber
import os
import json
import pandas as pd


folder_path = "data/"

for filename in os.listdir(folder_path):
    if filename.lower().endswith(".pdf"):
        file_path = os.path.join(folder_path, filename)
        brochure = "PSHB" if "PSHB" in filename else "FEHB"
        plan = "Blue-Focus" if "Blue-Focus" in filename else "Standard-Basic"
        print(f"Reading: {file_path}")
        rows = []
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                text = page.extract_text() or ""
                tables = page.extract_tables() or []
                
                content_in_tables =  "|".join([table_cell for table in tables for row in table for table_cell in row if table_cell]).split("|") if tables else ""

                rows.append({
                    "document_id": filename.replace(".pdf", ""),
                    "brochure": brochure,
                    "plan": plan, 
                    "page_number": page.page_number,
                    "text": text.strip(),
                    "text_length": len(text),
                    "num_tables": len(tables),
                    "tables": content_in_tables,  # keep raw for now
                })
        df = pd.DataFrame(rows)
        df.to_csv(f"{brochure + plan}.csv", index=False)
```
