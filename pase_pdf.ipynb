```python
# Clean PDF and store metadata files
import pdfplumber
import os
import json
import pandas as pd


rows = []

with pdfplumber.open("FEP-Blue-Focus-brochure-2026.pdf") as pdf:
    for page in pdf.pages:
        text = page.extract_text() or ""
        tables = page.extract_tables() or []
        
        
        content_in_tables =  "|".join([table_cell for table in tables for row in table for table_cell in row if table_cell]).split("|") if tables else ""

        rows.append({
            "document_id": "FEP_Blue_Focus_2026",
            "page_number": page.page_number,
            "text": text.strip(),
            "text_length": len(text),
            "num_tables": len(tables),
            "tables": content_in_tables,  # keep raw for now
        })

df = pd.DataFrame(rows)
```

```python
df.to_csv("FEP_Blue_Focus_2026_metadata.csv", index=False)
```
